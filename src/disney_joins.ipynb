```
    [TOP ( expression ) [PERCENT] [ WITH TIES ] ]   
	FROM table_source 
	[ WHERE search_condition ]
	[ GROUP BY group_by_expression ]
	[ HAVING search_condition ]
	[ ORDER BY order_expression [ ASC | DESC ] ]
```

````sql
select top 10 * from movie_gross
````

````sql
select top 5 * from director
````

````sql
select * 
    from movie_gross inner join director on 
        movie_gross.movie_title = director.movie_title
````

````sql
select movie_gross.movie_title, inflation_adjusted_gross, director
    from movie_gross inner join director on 
        movie_gross.movie_title = director.movie_title
````

````sql
select m.movie_title, inflation_adjusted_gross, director
    from movie_gross as m inner join director as d on 
        m.movie_title = d.movie_title
````

````sql
select m.movie_title, inflation_adjusted_gross, director
    from movie_gross as m left join director as d on 
        m.movie_title = d.movie_title
````

````sql
select m.movie_title, inflation_adjusted_gross, director
    from movie_gross as m right join director as d on 
        m.movie_title = d.movie_title
````

````sql
select m.movie_title, inflation_adjusted_gross, director
    from director as d right join movie_gross as m on 
        m.movie_title = d.movie_title
````

````sql
select * from disney_char
````

````sql
select m.movie_title, inflation_adjusted_gross, director, song
    from movie_gross as m left join director as d on 
        m.movie_title = d.movie_title left join disney_char as c 
            on m.movie_title = c.movie_title
````

````sql
with cte as (
    select m.movie_title, inflation_adjusted_gross, director, song
        from movie_gross as m left join director as d on 
            m.movie_title = d.movie_title left join disney_char as c 
                on m.movie_title = c.movie_title
)
select director, sum(inflation_adjusted_gross) [total gross] from cte 
    group by director
    having director is not NULL
    order by sum(inflation_adjusted_gross) desc
````
